- простые графики 

```python
# имитация данных
# массив (вектор-строка) из 1000 значений
# линейной последовательности от 0 до 10
x = np.linspace(0, 10, 1000)
x
```

```python
# данные для оси y
y = np.sin(x)
```

```python
# рисуем график
plt.plot(x, y)
plt.show()
```

```python
# совмещение двух графиков в одном окне
plt.plot(x, y)
plt.plot(x, np.cos(x))
plt.show()
```

```python
# интерактивный режим
%matplotlib notebook

# метки и легенды графиков

# легенды
plt.plot(x, y, label='sin')
plt.plot(x, np.cos(x), label='cos')

plt.legend()

# заголовок 
plt.title("Синус и косинус от x")
# метки 
plt.xlabel("x")
plt.ylabel("sin(x) & cos(x)")   
plt.show()
```

```python
# цвет и вид линий

# цвета можно задавать разными способами
plt.plot(x, np.sin(x), color='blue') #по названию
plt.plot(x, np.sin(x + 1), color='g') #по перечню букв цветов (rgbcmyk)
plt.plot(x, np.sin(x + 2), color='0.75') # оттенки серого в диапазоне от 0 до 1 
plt.plot(x, np.sin(x + 3), color=(1.0, 0.5, 0.2)) # RGB-кортеж, значения от 0 до 1
plt.show()
```

```python
# вид линий
plt.plot(x, x + 3, linestyle='--') # штриховая линия
plt.plot(x, x + 2, linestyle=':') # пунктирная линия
plt.plot(x, x + 1, linestyle='-.') # штрихпунктирная линия


plt.show()
```

```python
# объединение кодов вида и цвета линий
plt.plot(x, x + 3, '--g') # штриховая зеленая линия
plt.plot(x, x + 2, ':k') # пунктирная линия
plt.plot(x, x + 1, '-.r') # штрихпунктирная красная линия
plt.show()
```

- Простые диаграммы рассеивания

```python
# имитационные данные
x = np.random.randn(50)
y = np.random.randn(50)

colors = np.random.randn(50)
sizes = np.random.randn(50) * 100

# диаграмма рассеивания
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.colorbar(label="colors")
plt.show()
```

- Пример с городами Якутии

```python
# загрузка Excel-таблицы
cities = pd.read_excel('cities_of_Yakutia.xlsx')
cities
```

```python
# извлечение данных 
pop = cities['население']
area = cities['площадь']

lat = cities['широта']
lon = cities['долгота']

# lon
```

```python
# визуализация 

pop_log = np.log10(pop)

plt.scatter(lon, lat, c=pop_log, s=area)
plt.colorbar()

plt.title('Города Якутии')
plt.xlabel('долгота')
plt.yxlabel('широта')

plt.show()
```
